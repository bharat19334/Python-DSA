# Create Target Array in the Given Order
## Problem Statement
You are given two arrays: nums and index.

- nums contains the values.
- index tells where to put each value.
Start with an empty target array.

For each value in nums:
- Put the value at the position given in index.
- If a value is already there, move it and the remaining values one step to the right.
After adding all values, return the target array.

## Approach

We will create an empty target array.

For each iteration:
- Take the value nums[i].
- Take the position index[i].
- Insert the value at the given position in the target array.
- If an element already exists at that position, shift that element and all subsequent elements one position to the right.
- Place the new value at the required position.
After processing all elements, return the final target array

Step 1:
Insert 0 at index 0
Target = [0]

Step 2:
Insert 1 at index 1
Target = [0,1]

Step 3:
Insert 2 at index 2
Target = [0,1,2]

Step 4:
Insert 3 at index 2
Target = [0,1,3,2]

Step 5:
Insert 4 at index 1
Target = [0,4,1,3,2]

```python
def createTargetArray(nums, index):
    i = 0
    ans = []
    while i<=len(nums)-1:
        ans.insert(index[i],nums[i])
        i+=1
    return ans
nums = [0,1,2,3,4]
index = [0,1,2,2,1]
print(createTargetArray(nums,index))
```
